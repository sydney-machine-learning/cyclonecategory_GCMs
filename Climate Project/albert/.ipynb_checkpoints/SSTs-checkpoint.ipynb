### South Indian Ocean

```python
# It is too computationally demanding to create a big old region so I will split it up into 
# little chunks around areas of interest in different regions of the South Indian Ocean

# WHOLE DATA RANGE IS longitude: 30:134.6, -latitude: 50:-0.3125]

# look at squares of size 5
square_size = 5 

# TODO: Show these regions on a map

locations = [
    {"longitude": 55.601746, "latitude": -3.764745}, # Seychelles
    {"longitude": 56.47, "latitude": -20.45}, # Mauritius and Reunion
    {"longitude": 73.108510, "latitude": -3.787101}, # south of Maldives
#     {"longitude": 104.677851, "latitude": -6.608402}, # west coast of Java -- this one didnt have data
    {"longitude": 126.852503, "latitude": -11.057182} # Timor Sea north of Darwin
]
    
# si_north_lat =  -10
# si_south_lat = -20
# si_west_long = 80
# si_east_long = 90

south_indian_ssts = [daily_sst_dataset.sel(
    latitude=  slice(location["latitude"]  - square_size/2, location["latitude"]  + square_size/2), 
    longitude= slice(location["longitude"] - square_size/2, location["longitude"] + square_size/2)
) for location in locations]

south_indian_ssts
```

```python
mean_ssts = [sst_dataset['tos_mean_mean'] for sst_dataset in south_indian_ssts] # DAILY MEAN???
mean_ssts 
```

```python
# create rolling mean dataframe to smooth our data (a bit crude)
rolling_avg_ssts = [mean_sst_dataset.rolling(time=365, center=True).mean().dropna('time') for mean_sst_dataset in mean_ssts]
[sst['time'] for sst in rolling_avg_ssts]
```

```python
# select datapoints in the Southern Hemisphere cyclone season
cyclone_szn_rolling = [rolling_array.sel(time=rolling_array.time.dt.month.isin([12,1,2,3])) for rolling_array in rolling_avg_ssts]
cyclone_szn_rolling
```

```python
# THIS IS VERY INEFFICIENT AND EATING UP MY MEMORY -- IS THERE A BETTER APPROACH?????
# I might try porting this over to colab
dfs = []
for rolling_sst_array in cyclone_szn_rolling:

    longitudes = rolling_sst_array['longitude']
    latitudes = rolling_sst_array['latitude']


    rolling_sst_array['time'] = rolling_sst_array.indexes['time'].to_datetimeindex()

    # convert to a dataframe for plotting....
    df = None
    for longitude in longitudes:
        for latitude in latitudes:
            time_series = rolling_sst_array.sel(longitude=longitude, latitude=latitude, method='nearest').squeeze(drop=True).to_dataframe()
            if df is None:
                df = time_series
            else:
                df = pd.concat([df, time_series]) 
    dfs.append(df)

dfs
```

```python
sns.relplot(data=dfs[0], x="time", y="tos_mean_mean", kind='line', errorbar="sd").set(title='Seychelles')
 
```

```python
sns.relplot(data=dfs[1], x="time", y="tos_mean_mean", kind='line', errorbar="sd").set(title='Mauritius and Reunion')
```

```python
sns.relplot(data=dfs[2], x="time", y="tos_mean_mean", kind='line', errorbar="sd").set(title='South of Maldives') 
```

```python
sns.relplot(data=dfs[3], x="time", y="tos_mean_mean", kind='line', errorbar="sd").set(title='Timor Sea')
```
