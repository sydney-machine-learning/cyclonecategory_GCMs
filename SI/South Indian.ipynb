```python
%matplotlib inline
import numpy as np
import pandas as pd
import xarray as xr
from datetime import datetime,timedelta
from matplotlib import pyplot as plt
from netCDF4 import Dataset
```

```python
ds=Dataset(r'C:\Users\CHAITANYA AGRAWAL\OneDrive - IIT Delhi\Desktop\UNSW\SI_SST.nc')
url_1="https://raw.githubusercontent.com/sydney-machine-learning/cyclonedatasets/main/SouthIndian-SouthPacific-Ocean/South_indian_hurricane.csv"
df_1 = pd.read_csv(url_1, sep=',',error_bad_lines=False)
```

```python
df_1['Time']=pd.to_datetime(df_1['Time'],format='%Y%m%d%H')
time=ds.variables['time'][:]
lat=ds.variables['latitude'][:]
lon=ds.variables['longitude'][:]
sst=ds.variables['sst']
```

```python
intial=1900010100
initial_date=pd.to_datetime(intial,format='%Y%m%d%H')
df_1["diff"]=df_1['Time']-initial_date
df_1['total_1']=(df_1['diff']/timedelta(seconds=1))/3600
df_1['sst_1']=""
df_1
```

```python
for index,row in df_1.iterrows():
    latitude=row['lat_tenth']
    longitude=row['lon_tenth']
    total_1=row['total_1']

    
    diff_lat=(lat-latitude)**2
    diff_lon=(lon-longitude)**2
    diff_time=(time-total_1)**2

    
    min_lat=diff_lat.argmin()
    min_lon= diff_lon.argmin()
    min_time=diff_time.argmin()

    
    df_1.loc[index,'sst_1']=sst[min_time,0,min_lat,min_lon]
 
```

```python
df_1.to_csv('Subs_SI.csv',index=False)
with open('Subs_SI.csv','r') as in_file, open('Output_SI.csv','w') as out_file:
  
    seen = set() 
    
    for line in in_file:
        if line in seen: 
          continue 

        seen.add(line)
        out_file.write(line)
```

```python
df_1 = pd.read_csv('Output_SI.csv')
df_1
```

```python
for index,row in df_1.iterrows():
        start=df_1['total_1'][index]
        for days in range(1,31):
            start=start-24*days

            latitude=df_1['lat_tenth'][index]
            longitude=df_1['lon_tenth'][index]

            diff_lat=(lat-latitude)**2
            diff_lon=(lon-longitude)**2
            diff_time=(time-start)**2


            min_lat=diff_lat.argmin()
            min_lon= diff_lon.argmin()
            min_time=diff_time.argmin()



            df_1.loc[index,days]=sst[min_time,0,min_lat,min_lon]   
    
```
