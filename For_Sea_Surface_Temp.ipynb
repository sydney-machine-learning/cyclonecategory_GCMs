```python
%matplotlib inline
import numpy as np
import pandas as pd
import xarray as xr
from datetime import datetime,timedelta
from matplotlib import pyplot as plt
from netCDF4 import Dataset
```

```python
ds=Dataset(r'C:\Users\CHAITANYA AGRAWAL\OneDrive - IIT Delhi\Desktop\UNSW\SP_SST.nc')
url="https://raw.githubusercontent.com/sydney-machine-learning/cyclonedatasets/main/SouthIndian-SouthPacific-Ocean/South_pacific_hurricane.csv"
df = pd.read_csv(url, sep=',',error_bad_lines=False)
```

```python
time=ds.variables['time'][:]
lat=ds.variables['latitude'][:]
lon=ds.variables['longitude'][:]
sst=ds.variables['sst']
ds['time']
```

```python
df['Time']=pd.to_datetime(df['Time'],format='%Y%m%d%H')
```

```python
intial=1900010100
initial_date=pd.to_datetime(intial,format='%Y%m%d%H')
df["diff"]=df['Time']-initial_date
```

```python
df
```

```python
df['total']=(df['diff'][0]/timedelta(seconds=1))/3600
df['sst']=""

```

```python
sst=ds.variables['sst'][:]
```

```python
sst
```

```python
for index,row in df.iterrows():
    latitude=row['lat_tenth']
    longitude=row['lon_tenth']
    total=row['total']
    
    diff_lat=(lat-latitude)**2
    diff_lon=(lon-longitude)**2
    diff_time=(time-total)**2
    
    
    min_lat=diff_lat.argmin()
    min_lon= diff_lon.argmin()
    min_time=diff_time.argmin()
    
    
    df.loc[index,'sst']=sst[min_time,0,min_lat,min_lon]

    
```
